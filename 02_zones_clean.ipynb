```python
#last_wts_flow_zone_cleaned.csv
zones = pd.read_csv(r"C:\Users\hub_stag6\Documents\Data\last_wts_flow_zone_cleaned.csv", sep=",")
zones.tail(4)
```

```python
# Erase 'CF' of the sensor name 
zones.insert(1,'nom_capteur2', (zones["nom_capteur"].str.replace("^CF\s-","", regex=True)))
# Split the sensor name and the shop
zones[['name','shop']] = zones.nom_capteur2.str.split(" - ", expand = True)
# Sort DataFrame by zone
zones = zones.sort_values(by='zone_id').reset_index(drop=True)
# Capitalize the first letter of each word of the Quartier variable
zones.Quartier = zones.Quartier.str.title()
```

```python
zones = zones.loc[:, ['zone_id', 'nom_capteur', 'Quartier', 'shop', 'id _capteur_quartier',
       'type_capteur', 'address', 'zipcode', 'city', 'longitude','latitude']]
zones.tail(4)
```
